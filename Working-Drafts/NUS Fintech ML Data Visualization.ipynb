```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks_data = pd.read_csv("Project/DJIA_table.csv")
stocks_data.head()
```

```python
stocks_data.info()
```

```python
stocks_data['Date'] =  pd.to_datetime(stocks_data['Date'], format='%Y-%m-%d')
stocks_data['Date'].head()
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(20,10)})

adj_close = sns.lineplot(x="Date", y="Adj Close", data=stocks_data, linewidth=1, label = 'Adjusted Close')
adj_close.axes.set_title("Stock Performance from 2008 to 2017",fontsize=20)
adj_close.set_xlabel("Date", fontsize=15)
adj_close.set_ylabel("USD", fontsize=15)
adj_close.legend()

plt.show()
```

```python
latest_stocks = stocks_data[stocks_data['Date'] >= '2015-01-01']
latest_stocks.head(3)
```

```python
high_low = sns.lineplot(x="Date", y="High", data=latest_stocks, color = 'green', linewidth=0.7, label = 'Adjusted Close')
high_low = sns.lineplot(x="Date", y="Low", data=latest_stocks, color = 'red', linewidth=0.7, label = 'Adjusted Close')
high_low.axes.set_title("High-Low from 2015 onwards",fontsize=20)
high_low.set_xlabel("Date", fontsize=15)
high_low.set_ylabel("USD", fontsize=15)
high_low.legend()

plt.show()
```

```python
stocks_data['Volatility'] = stocks_data['High'] - stocks_data['Low']
latest_stocks['Volatility'] = latest_stocks['High'] - latest_stocks['Low']
```

```python
adj_close = sns.lineplot(x="Date", y="Volume", data=latest_stocks, linewidth=1, label = 'Volatility')
adj_close.axes.set_title("Volatility from 2015 to 2017",fontsize=20)
adj_close.set_xlabel("Date", fontsize=15)
adj_close.set_ylabel("USD", fontsize=15)
adj_close.legend()

plt.show()
```

```python
adj_close = sns.lineplot(x="Date", y="Adj Close", data=latest_stocks, color = 'orange', linewidth=1, label = 'Adj Close')
adj_close.axes.set_title("Volatility from 2015 to 2017",fontsize=20)
adj_close.set_xlabel("Date", fontsize=15)
adj_close.set_ylabel("USD", fontsize=15)
adj_close.legend()

plt.show()
```
